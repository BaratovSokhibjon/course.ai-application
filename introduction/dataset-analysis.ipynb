## Descriptive Analytics
Descriptive statistics provide a summary of data features:
- Minimum (Min)
- Maximum (Max)
- Mean
- Median
- Standard Deviation (SD)
- Range

```python
import numpy as np
import pandas as pd

data = [5, 8, 12, 12, 20]
series = pd.Series(data)

print("Minimum:", series.min())
print("Maximum:", series.max())
print("Mean:", series.mean())
print("Median:", series.median())
print("Standard Deviation:", series.std())
print("Range:", series.max() - series.min())
```

## Data Distribution & Visualization
Visualizations help us understand data distribution:
- Boxplot
- Histogram
- Line Chart

```python
import matplotlib.pyplot as plt

data = [5, 8, 12, 12, 20]

# Boxplot
plt.figure(figsize=(6,4))
plt.boxplot(data)
plt.title("Boxplot of Dataset")
plt.show()

# Histogram
plt.figure(figsize=(6,4))
plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title("Histogram of Dataset")
plt.show()

# Line chart
plt.figure(figsize=(6,4))
plt.plot(data, marker='o', color='green')
plt.title("Line Chart of Dataset")
plt.show()
```

## Data Correlation
- Correlation shows the relationship between two variables
- Range: -1 (negative) to +1 (positive)
- Tools: Scatter plot, Heatmap, Line chart

```python
import seaborn as sns

data = pd.DataFrame({
    "Study_Hours": [2, 4, 6, 8, 10],
    "Exam_Scores": [50, 60, 65, 70, 85]
})

# Scatter plot
plt.scatter(data["Study_Hours"], data["Exam_Scores"], color="orange")
plt.title("Study Hours vs Exam Scores")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.show()

# Correlation Heatmap
sns.heatmap(data.corr(), annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()

# Line chart
plt.plot(data["Study_Hours"], label="Study Hours", marker="o")
plt.plot(data["Exam_Scores"], label="Exam Scores", marker="x")
plt.legend()
plt.title("Line Chart of Study Hours & Exam Scores")
plt.show()
```

## Handling Duplicates and Missing Values
- Drop duplicates
- Handle missing numeric values (mean, median, mode)
- Handle missing categorical values (mode, ML imputation)

```python

# Example with artificial dataset
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Carol", "Dave", "Eve"],
    "Age": [20, None, 19, 21, 22],
    "Grade": ["A", None, None, "B", "A"],
    "Score": [90, 85, 78, None, 88]
})

print("Initial missing values:\n", df.isnull().sum())

# Fill numeric with median
num_cols = df.select_dtypes(include=['int64','float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill categorical with mode
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("\nAfter cleaning:\n", df)
```
