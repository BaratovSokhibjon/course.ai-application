```python
# Import required libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import pearsonr
```

```python
# Dataset setup based on the provided table
hours_sleep = [4, 5, 6, 7, 8]
productivity_score = [50, 55, 60, 65, 80]

# Create a DataFrame for easier data manipulation
data = pd.DataFrame({
    'Hours_Sleep': hours_sleep,
    'Productivity_Score': productivity_score
})

print("Dataset Overview:")
print("=" * 30)
print(data)
print(f"\nNumber of data points: {len(data)}")
print(f"Sleep range: {min(hours_sleep)} - {max(hours_sleep)} hours")
print(f"Productivity range: {min(productivity_score)} - {max(productivity_score)} points")
```

```python
# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(hours_sleep, productivity_score, color='blue', s=100, alpha=0.7, edgecolors='black')

# Add trend line
z = np.polyfit(hours_sleep, productivity_score, 1)
p = np.poly1d(z)
plt.plot(hours_sleep, p(hours_sleep), "r--", alpha=0.8, linewidth=2, label=f'Trend line: y = {z[0]:.1f}x + {z[1]:.1f}')

plt.title('Hours of Sleep vs Productivity Score', fontsize=14, fontweight='bold')
plt.xlabel('Hours of Sleep', fontsize=12)
plt.ylabel('Productivity Score', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()

# Annotate each point
for i, (x, y) in enumerate(zip(hours_sleep, productivity_score)):
    plt.annotate(f'({x}, {y})', (x, y), xytext=(5, 5), textcoords='offset points', 
                fontsize=10, ha='left')

plt.tight_layout()
plt.show()

print("📊 Scatter Plot Observation:")
print("The scatter plot shows a clear positive relationship between hours of sleep and productivity score.")
```

```python
# Calculate correlation coefficient
correlation_coeff, p_value = pearsonr(hours_sleep, productivity_score)

print("Correlation Analysis Results:")
print("=" * 35)
print(f"Pearson Correlation Coefficient: {correlation_coeff:.4f}")
print(f"P-value: {p_value:.6f}")
print()

# Interpretation
print("Interpretation:")
if abs(correlation_coeff) >= 0.8:
    strength = "very strong"
elif abs(correlation_coeff) >= 0.6:
    strength = "strong"
elif abs(correlation_coeff) >= 0.4:
    strength = "moderate"
elif abs(correlation_coeff) >= 0.2:
    strength = "weak"
else:
    strength = "very weak"

direction = "positive" if correlation_coeff > 0 else "negative"
print(f"📈 There is a {strength} {direction} correlation between hours of sleep and productivity score.")

# Additional statistics
print(f"\n📊 Additional Statistics:")
print(f"Coefficient of determination (R²): {correlation_coeff**2:.4f}")
print(f"This means {correlation_coeff**2*100:.1f}% of the variance in productivity can be explained by hours of sleep.")
```

```python
# Create line chart for both variables on the same plot
plt.figure(figsize=(12, 8))

# Create x-axis values (data point indices)
x_values = list(range(1, len(hours_sleep) + 1))

# Plot both variables
plt.plot(x_values, hours_sleep, marker='o', linewidth=3, markersize=8, 
         label='Hours of Sleep', color='blue', linestyle='-')
plt.plot(x_values, productivity_score, marker='s', linewidth=3, markersize=8, 
         label='Productivity Score', color='red', linestyle='-')

plt.title('Hours of Sleep vs Productivity Score - Line Chart Comparison', fontsize=14, fontweight='bold')
plt.xlabel('Data Point Index', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)

# Add value labels on points
for i, (sleep, prod) in enumerate(zip(hours_sleep, productivity_score)):
    plt.annotate(f'{sleep}h', (x_values[i], sleep), xytext=(0, 10), 
                textcoords='offset points', ha='center', fontsize=10, color='blue')
    plt.annotate(f'{prod}', (x_values[i], prod), xytext=(0, -15), 
                textcoords='offset points', ha='center', fontsize=10, color='red')

plt.xticks(x_values)
plt.tight_layout()
plt.show()

print("📈 Line Chart Observation:")
print("Both variables show similar upward trends, confirming their positive correlation.")
print("The productivity score increases more steeply than hours of sleep, indicating")
print("that small increases in sleep can lead to significant productivity improvements.")
```
